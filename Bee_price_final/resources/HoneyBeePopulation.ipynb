```python
import pandas as pd
```

```python
all_data = "honeybee_by_state.csv"
df = pd.read_csv(all_data)
```

```python
df.head()
```

```python
simple_df = df[["Year", "Period", "State", "State ANSI", "Data Item", "Value"]]
```

```python
simple_df.head()
```

```python
market_year_df = simple_df.loc[simple_df["Period"] == "MARKETING YEAR", :]
```

```python
market_year_df.head()
```

```python
market_year_df.count()
```

```python
bee_pop_df = market_year_df[["Year","State","Value"]]
```

```python
bee_pop_df = bee_pop_df.rename(columns={"Value":"Population"})
```

```python
bee_pop_df.head()
```

```python
bee_pop_df.count()
```

```python
bee_pop_df.to_csv('bee_pop_clean.csv')
```
